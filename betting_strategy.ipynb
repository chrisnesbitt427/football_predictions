```python
path = r"C:\Users\cnesb\football_predictions\data\3_leagues_all_out2.csv"
df = pd.read_csv(path)
```

```python

df =df[['Home Side', 'Away Side', 'home_Gls', 'away_Gls', 'Predicted', 'win_loss', 'B365H', 'B365D', 'B365A',  'Certainty']]
new_df = df
new_df['Bookies Chance Home Win'] = new_df['B365H'] / (new_df['B365H'] + new_df['B365D'] + new_df['B365A'])
new_df['Bookies Chance Home Loss'] = new_df['B365A'] / (new_df['B365H'] + new_df['B365D'] + new_df['B365A'])

# Mapping 0 to 'Loss' and 1 to 'Win'
new_df['Predicted'] = new_df['Predicted'].map({0: 'Loss', 1: 'Win'})
new_df['win_loss'] = new_df['win_loss'].map({0: 'Loss', 1: 'Win'})
new_df['Bookies_Odds'] = new_df.apply(lambda row: row['Bookies Chance Home Win'] if row['Predicted'] == 'Win' else row['Bookies Chance Home Loss'], axis=1)
new_df
```

```python
new_df = new_df.drop(columns=['Bookies Chance Home Win', 'Bookies Chance Home Loss'])
new_df
```

```python
df = new_df

# Convert fractional odds to decimal odds (add 1)
df['Decimal_Odds'] = df['Bookies_Odds'] + 1

# Calculate the Kelly Criterion for each row
df['Kelly_Fraction'] = df.apply(
    lambda row: max(((row['Certainty'] * (row['Decimal_Odds'] - 1)) - (1 - row['Certainty'])) / (row['Decimal_Odds'] - 1), 0), axis=1
)

# Track the bankroll assuming an initial amount
initial_bankroll = 1000
df['Bet_Size'] = df['Kelly_Fraction'] * initial_bankroll

# Simulate the result of the bet
df['Profit_Loss'] = df.apply(
    lambda row: row['Bet_Size'] * (row['Decimal_Odds'] - 1) if row['Predicted'] == row['win_loss'] else -row['Bet_Size'], axis=1
)

# Calculate the updated bankroll after each bet
df['Bankroll'] = initial_bankroll + df['Profit_Loss'].cumsum()

print(df)
```

```python
df.to_csv("hi2.csv")
```
